# tkinter를 이용하여 버튼을 누를 때마다 6개의 랜덤 번호를 출력하는 코드 만들기

```python
#GUi 생성을위해 tkinter 라이브러리 임포트
import tkinter as tk
#tkinter의 폰트 수정을 위해 tkinter.font 라이브러리 임포트
import tkinter.font
#난수 생성을 위해 random 라이브러리 임포트
import random

#1~45사이 번호를 생성
lotto_num = range(1,46)

#버튼을 누르면 동작하는 함수 생성
def btn_click():
    #6개의 번호를 랜덤하게 출력(중복값 x)
    print(random.sample(lotto_num,6))

#GUI 설정
#윈도우 창 생성
window = tk.Tk()
#타이틀 변경
window.title("lotto")
#창의 크기와 최초 생성 위치 지정
window.geometry("400x200+800+300")
#창 크기 재조절 여부 설정
window.resizable(False,False)

#버튼 생성
button = tk.Button(window,overrelief="solid",text="번호확인",width = 15, 
                   command=btn_click,repeatdelay=1000,repeatinterval=100)
#버튼을 표시함
button.pack()
#윈도우가 닫히기 전까지 반복
window.mainloop()
```

## 같이 볼만한 개념

tk.Button()의 파라미터들 

overrelief: 버튼에 마우스를 올렸을 때 버튼의 테두리 모양, text:버튼에 표시할 문자열,
command: 사용자 정의함수를 실행,repeatdelay:버튼이 눌러진 상태에서 command 실행까지의 대기 시간,
repeatinterval:버튼이 눌러진 상태에서 command 실행의 반복 시간
