```python
import numpy as np
import cv2

#사진 그림 변환
ff = np.fromfile(r'C:\Users\user\사진을 그림으로 변환(Open CV)\dogimage.jpg',np.uint8)
img = cv2.imdecode(ff,cv2.IMREAD_UNCHANGED)
img = cv2.resize(img ,dsize=(0,0), fx=1.0,fy=1.0,interpolation = cv2.INTER_LINEAR)

def onChange(pos):
    pass

#트랙바 윈도우를 생성합니다.
cv2.namedWindow("Trackbar Windows")

#트랙의 최소값, 최대값을 설정합니다. 트랙이 움직일 때 마다 동작하는 함수를 지정합니다.
cv2.createTrackbar("sigma_s","Trackbar Windows",0,200,onChange)
cv2.createTrackbar("sigma_r","Trackbar Windows",0,100,onChange)

#트랙의 기본 위치를 지정합니다.(초기값 지정)
cv2.setTrackbarPos("sigma_s","Trackbar Windows",100)
cv2.setTrackbarPos("sigma_r","Trackbar Windows",10)

#실행을 위한 반복문
while True:
    
    #100ms간 입력이 없으면 timeout으로 이 코드를 종료하고 다음 코드를 실행합니다. q의 키값이 입력되면 break로 while문을 종료합니다.
    if cv2.waitKey(100) == ord('q'):
        break
    
    #트랙의 포지션으로 sigma_s_value, sigma_r_value 값을 받습니다. sigma_r_value값은 나누기 100을 합니다.
    sigma_s_value = cv2.getTrackbarPos("sigma_s","Trackbar Windows")
    sigma_r_value = cv2.getTrackbarPos("sigma_r","Trackbar Windows")/100.0
    
    print("sigma_s_value",sigma_s_value)
    print("sigma_r_value",sigma_r_value)
    
    #트랙의 포지션에 따라 이미지를 그림화합니다.
    cartoon_img = cv2.stylization(img,sigma_s = sigma_s_value, sigma_r=sigma_r_value) #sigma_s값과 sigma_r값을 조절하여 이미지를 그림화합니다,

    cv2.imshow('cartoon view',cartoon_img)

#opencv의 모든창이 닫히고 종료됩니다. 
cv2.destroyAllWindows()
```

sigma_s = 이미지가 부드럽게 되는 정도를 제어합니다. 값이 클수록 이미지가 더 부드러워지지만 계산 속도도 느려집니다.

sigma_r =이미지를 부드럽게 하는 동안 가장자리를 유지하려는 경우 중요합니다. 작으면 sigma_r이 매우 유사한 색상만 평균화(즉, 매끄럽게 됨)되는 반면 많이 다른 색상은 그대로 유지됩니다.
