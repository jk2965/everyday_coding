```python
import tkinter as tk
import math
import numpy


calc = tk.Tk() #윈도우 창 생성

calc.title("계산기") #타이틀 정함 
calc.geometry("800x600+100+100") #윈도우창의 너비x윈도우창의 높이(곱하기가 아니라 x로 쳐야됨) + 초기 x좌표 + 초기 y좌표 지정
calc.resizable(0,0) #윈도우창 사이즈 조절 불가하게 설정

upper_frame = tk.Frame(calc, width=800, height=70) #넓이 400 높이 70을 가지는 프레임 생성
upper_frame.pack(pady=40) #x간격 = 40

down_frame = tk.Frame(calc, width=600, height=100) #넓이 400 높이 100을 가지는 프레임 생성
down_frame.pack(padx=10, pady=10) #x간격 = 10 ,y간격 = 10

entry = tk.Entry(upper_frame, width=40, font=("Courier",18), borderwidth=5) #텍스트 출력을 위한 기입창 생성
entry.pack() #엔트리 위젯을 배치
entry.insert(0,"") #기입창의 0번째 인덱스에 "" 입력

def button_clicked(number):
    current = entry.get() #기입창의 정보를 받아옴
    entry.delete(0, tk.END) #기입창의 0번 인덱스를 지움 (기입창 초기화)
    entry.insert(0,  str(current) + str(number)) #기입창의 0번 자리에 입력되어있던 전체 기입창의 문자와 눌려진 문자를 입력 
    
def button_clear():
    entry.delete(0, tk.END) # 기입창의 0번 인덱스를 지움 (기입창 초기화)

def button_add():
    first_number = entry.get() #기입창의 정보를 받아옴
    global num1 #num1 전역변수 선언
    global math #math 전역변수 선언
    math = 'add' #math 전역변수에 값 할당
    num1 = float(first_number) #num1 전역변수에 first_number 값 할당
    entry.delete(0, tk.END) # 기입창의 0번 인덱스를 지움 (기입창 초기화)

def button_sub():
    first_number = entry.get() #기입창의 정보를 받아옴
    global num1 #num1 전역변수 선언
    global math #math 전역변수 선언
    math = 'subtract' #math 전역변수에 값 할당
    num1 = float(first_number) #num1 전역변수에 first_number 값 할당
    entry.delete(0, tk.END) # 기입창의 0번 인덱스를 지움 (기입창 초기화)

def button_mul():
    first_number = entry.get() #기입창의 정보를 받아옴
    global num1 #num1 전역변수 선언
    global math #math 전역변수 선언
    math = 'multiply' #math 전역변수에 값 할당
    num1 = float(first_number) #num1 전역변수에 first_number 값 할당
    entry.delete(0, tk.END) # 기입창의 0번 인덱스를 지움 (기입창 초기화)

def button_div():
    first_number = entry.get() #기입창의 정보를 받아옴
    global num1 #num1 전역변수 선언
    global math #math 전역변수 선언
    math = 'devide' #math 전역변수에 값 할당
    num1 = float(first_number) #num1 전역변수에 first_number 값 할당
    entry.delete(0, tk.END) #기입창의 0번 인덱스를 지움 (기입창 초기화)
    
def button_abs():
    first_number = entry.get() #기입창의 정보를 받아옴
    global num1 #num1 전역변수 선언
    global math #math 전역변수 선언
    math = 'abs' #math 전역변수에 값 할당
    num1 = float(first_number) #num1 전역변수에 first_number 값 할당
    entry.delete(0, tk.END) #기입창의 0번 인덱스를 지움 (기입창 초기화)
    entry.insert(0, abs(num1))

def button_pow():
    first_number = entry.get() #기입창의 정보를 받아옴
    global num1 #num1 전역변수 선언
    global math #math 전역변수 선언
    math = 'power' #math 전역변수에 값 할당
    num1 = float(first_number) #num1 전역변수에 first_number 값 할당
    entry.delete(0, tk.END) #기입창의 0번 인덱스를 지움 (기입창 초기화)
    
def button_sqrt():
    first_number = entry.get() #기입창의 정보를 받아옴
    global num1 #num1 전역변수 선언
    try:
        num1 = float(first_number) #num1 전역변수에 first_number 값 할당
        entry.delete(0, tk.END) #기입창의 0번 인덱스를 지움 (기입창 초기화)
        log1 = math.sqrt(num1)
        entry.insert(0, log1)
    except:
        entry.delete(0, tk.END) #기입창의 0번 인덱스를 지움 (기입창 초기화)
        entry.insert(0, "음수나 0, 문자열은 자연로그 값을 구할 수 없습니다.")
    
def button_ln():
    first_number = entry.get() #기입창의 정보를 받아옴
    global num1 #num1 전역변수 선언
    num1 = float(first_number) #num1 전역변수에 first_number 값 할당
    entry.delete(0, tk.END) #기입창의 0번 인덱스를 지움 (기입창 초기화)
    try:
        log1 = math.log(num1)
        entry.insert(0, log1)
    except:
        entry.insert(0, "음수나 0, 문자열은 자연로그 값을 구할 수 없습니다.")
        
def button_log10():
    first_number = entry.get() #기입창의 정보를 받아옴
    global num1 #num1 전역변수 선언
    num1 = float(first_number) #num1 전역변수에 first_number 값 할당
    entry.delete(0, tk.END) #기입창의 0번 인덱스를 지움 (기입창 초기화)
    try:
        log1 = math.log10(num1)
        entry.insert(0, log1)
    except:
        entry.insert(0, "음수나 0, 문자열은 상용로그 값을 구할 수 없습니다.")

def button_exp():
    first_number = entry.get() #기입창의 정보를 받아옴
    global num1 #num1 전역변수 선언

    try:
        num1 = float(first_number) #num1 전역변수에 first_number 값 할당
        entry.delete(0, tk.END) #기입창의 0번 인덱스를 지움 (기입창 초기화)
        exp1 = math.exp(num1)
        entry.insert(0, exp1)
    except:
        entry.delete(0, tk.END) #기입창의 0번 인덱스를 지움 (기입창 초기화)
        entry.insert(0, "문자열은 지수함수 값을 구할 수 없습니다.")


def button_equal():
    num2 = entry.get() #기입창의 정보를 받아옴
    entry.delete(0, tk.END) # 기입창의 0번 인덱스를 지움 (기입창 초기화)
    
    if math == 'add':
        if num2 == "":
            entry.insert(0,"더할 값을 입력하세요.")
        else:
            entry.insert(0,num1 + float(num2)) #기입창의 0번 인덱스에 num1의 값과 num2의 값을 더한 값을 입력

    if math == 'subtract':
        if num2 == "":
            entry.insert(0,"뺄 값을 입력하세요.")
        else:
            entry.insert(0,num1 - float(num2)) #기입창의 0번 인덱스에 num1의 값과 num2의 값을 뺀 값을 입력

    if math == 'multiply':
        if num2 == "":
            entry.insert(0,"곱할 값을 입력하세요.")
        else:
            entry.insert(0,num1 * float(num2)) #기입창의 0번 인덱스에 num1의 값과 num2의 값을 곱한 값을 입력

    if math == 'devide':
        try:
            entry.insert(0,num1 / float(num2))
        
        except:
            entry.insert(0,"수식이 잘못되었습니다.")
        
    if math == 'power':
        if num2 == "":
            entry.insert(0,"곱할 값을 입력하세요.")
        else:
            entry.insert(0,num1 ** float(num2)) #기입창의 0번 인덱스에 num1의 값과 num2의 값을 곱한 값을 입력

btn7 = tk.Button(down_frame,text='7', padx=15, pady=10, font=("Courier",15),command=lambda: button_clicked(7))
btn7.grid(column=0, row=0, padx=5, pady=5)
btn8 = tk.Button(down_frame,text='8', padx=15, pady=10, font=("Courier",15),command=lambda: button_clicked(8))
btn8.grid(column=1, row=0, padx=5, pady=5)
btn9 = tk.Button(down_frame,text='9', padx=15, pady=10, font=("Courier",15),command=lambda: button_clicked(9))
btn9.grid(column=2, row=0, padx=5, pady=5)

btn4 = tk.Button(down_frame,text='4', padx=15, pady=10, font=("Courier",15),command=lambda: button_clicked(4))
btn4.grid(column=0, row=1, padx=5, pady=5)
btn5 = tk.Button(down_frame,text='5', padx=15, pady=10, font=("Courier",15),command=lambda: button_clicked(5))
btn5.grid(column=1, row=1, padx=5, pady=5)
btn6 = tk.Button(down_frame,text='6', padx=15, pady=10, font=("Courier",15),command=lambda: button_clicked(6))
btn6.grid(column=2, row=1, padx=5, pady=5)


btn1 = tk.Button(down_frame,text='1', padx=15, pady=10, font=("Courier",15),command=lambda: button_clicked(1))
btn1.grid(column=0, row=2, padx=5, pady=5)
btn2 = tk.Button(down_frame,text='2', padx=15, pady=10, font=("Courier",15),command=lambda: button_clicked(2))
btn2.grid(column=1, row=2, padx=5, pady=5)
btn3 = tk.Button(down_frame,text='3', padx=15, pady=10, font=("Courier",15),command=lambda: button_clicked(3))
btn3.grid(column=2, row=2, padx=5, pady=5)

btn_pm = tk.Button(down_frame,text='+/-', padx=5, pady=10, font=("Courier",15),command=lambda: button_clicked('-'))
btn_pm.grid(column=0, row=3, padx=5, pady=5)
btn0 = tk.Button(down_frame,text='0', padx=15, pady=10, font=("Courier",15),command=lambda: button_clicked(0))
btn0.grid(column=1, row=3, padx=5, pady=5)
btn_p = tk.Button(down_frame,text='.', padx=15, pady=10, font=("Courier",15),command=lambda: button_clicked('.'))
btn_p.grid(column=2, row=3, padx=5, pady=5)


btn_mul = tk.Button(down_frame,text='×', padx=15, pady=10, font=("Courier",15),command=button_mul, bg='orange')
btn_mul.grid(column=3, row=0, padx=5, pady=5)
btn_sub = tk.Button(down_frame,text='-', padx=15, pady=10, font=("Courier",15),command=button_sub, bg='orange')
btn_sub.grid(column=3, row=1, padx=5, pady=5)
btn_add = tk.Button(down_frame, text='+', padx=15, pady=10, font=("Courier",15),command=button_add, bg='orange')
btn_add.grid(column=3, row=2, padx=5, pady=5)
btn_div = tk.Button(down_frame, text='÷', padx=15, pady=10, font=("Courier",15),command=button_div, bg='orange')
btn_div.grid(column=3, row=3, padx=5, pady=5)

btn_exp = tk.Button(down_frame, text='e', padx=15, pady=10, font=("Courier",15),command=button_exp, bg='orange')
btn_exp.grid(column=0, row=4, padx=5, pady=5)
btn_sqrt = tk.Button(down_frame, text='√', padx=15, pady=10, font=("Courier",15),command=button_sqrt, bg='orange')
btn_sqrt.grid(column = 1, row=4, padx=5, pady=5)
btn_c = tk.Button(down_frame, text='C', padx=15, pady=10, font=("Courier",15),command=button_clear, bg='orange')
btn_c.grid(column=2, row=4, padx=5, pady=5)
btn_res = tk.Button(down_frame, text='=', padx=15, pady=10, font=("Courier",15),command=button_equal, bg='orange')
btn_res.grid(column=3, row=4, padx=5, pady=5)



btn_abs = tk.Button(down_frame, text='abs', padx=15, pady=10, font=("Courier",15),command=button_abs, bg='orange')
btn_abs.grid(column = 4, row=0, padx=5, pady=5)
btn_pow = tk.Button(down_frame, text='pow', padx=15, pady=10, font=("Courier",15),command=button_pow, bg='orange')
btn_pow.grid(column = 4, row=1, padx=5, pady=5)
btn_ln = tk.Button(down_frame, text='ln', padx=15, pady=10, font=("Courier",15),command=button_ln, bg='orange', width = 4)
btn_ln.grid(column = 4, row=2, padx=5, pady=5)
btn_log10 = tk.Button(down_frame, text='log10', padx=15, pady=10, font=("Courier",15),command=button_log10, bg='orange',width=4)
btn_log10.grid(column = 4, row=3, padx=5, pady=5)

calc.mainloop() #윈도우가 종료 될 때 까지 실행
```
