```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_lfw_people
from sklearn.model_selection import train_test_split
import mglearn
```

```python
#데이터 수집
people = fetch_lfw_people(min_faces_per_person=20,resize=0.7)
image_shape=people.images[0].shape
```

```python
fig,axes = plt.subplots(2,5,figsize=(15,8),
                       subplot_kw={'xticks':(),'yticks':()})

for target,image,ax in zip(people.target,people.images,axes.ravel()):
    ax.imshow(image)
    ax.set_title(people.target_names[target])
```

```python
print('people',people.images.shape)
print('클래스 개수:',len(people.target_names))
```

```python
# 각 타깃을 나타난 횟수
counts = np.bincount(people.target)
counts
```

```python
#타깃별 이름과 횟수 출력
for i,(count,name) in enumerate(zip(counts,people.target_names)):
    print('{0:25}{1:3}'.format(name,count),end='    ')
    if (i+1) % 3 == 0 :
        print()
```

```python
#스케일 조정
mask= np.zeros(people.target.shape,dtype=bool)
for target in np.unique(people.target):
    mask[np.where(people.target==target)[0][:50]]=1
    
x_people = people.data[mask]
y_people = people.target[mask]

x_people = x_people/255 
#흑백이미지는 0에서 255사이 픽셀을 가지고 있기 때문에 
#0과 1사이로 조절을 위해 나눠줌
```

```python
from sklearn.neighbors import KNeighborsClassifier

#훈련,테스트 데이터 분할
x_train,x_test,y_train,y_test=train_test_split(x_people,y_people,
                                              stratify=y_people,
                                              random_state=0)
#knn모델 정의
knn = KNeighborsClassifier(n_neighbors=3)
#모델 학습
knn.fit(x_train,y_train)
#분류모델이므로 model.score()의 기준은 Accuracy
print('훈련 데이터 정확도:',knn.score(x_train,y_train))
print('테스트 데이터 정확도:',knn.score(x_test,y_test))
```

```python
pca= PCA(n_components=100,whiten=True,random_state=0)
pca.fit(x_train)
#pca모델 선정,학습
```

```python
x_train_pca = pca.transform(x_train)
x_test_pca = pca.transform(x_test)
#차원 축소
```

```python
x_train_pca.shape
#100차원으로 차원 축소됨
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train_pca,y_train)
#모델 선정,학습
```

```python
#분류모델이므로 model.score()의 기준은 Accuracy
print('훈련 데이터 정확도:',knn.score(x_train_pca,y_train))
print('테스트 데이터 정확도:',knn.score(x_test_pca,y_test))
```

```python
fig,axes = plt.subplots(3,5,figsize=(15,12),
                       subplot_kw={'xticks':(),'yticks':()})

for i,(component,ax) in enumerate(zip(pca.components_,axes.ravel())):
    ax.imshow(component.reshape(image_shape),cmap='viridis')
    ax.set_title('주성분 {}'.format(i+1))
```
