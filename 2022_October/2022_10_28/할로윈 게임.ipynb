```python
import pygame
import random

# 화면 크기 정보
SCREEN_WIDTH = 480
SCREEN_HEIGHT = 680

# 파이게임 초기화 및 화면 크기 지정
pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# 이미지 로드
background_image = pygame.image.load('gameimages/background.jpg')
player_image = pygame.image.load('gameimages/ghost.png')
#호박 이미지
pumpkin_image1 = pygame.image.load('gameimages/pumpkin1.jpg') 
pumpkin_image2 = pygame.image.load('gameimages/pumpkin2.jpg')
pumpkin_image3 = pygame.image.load('gameimages/pumpkin3.jpg')
pumpkin_image4 = pygame.image.load('gameimages/pumpkin4.jpg')
pumpkin_image5 = pygame.image.load('gameimages/pumpkin5.jpg')
#사탕 이미지
candy_image1 = pygame.image.load('gameimages/candy1.png')
candy_image2 = pygame.image.load('gameimages/candy2.jpg')
candy_image3 = pygame.image.load('gameimages/candy3.jpg')
candy_image4 = pygame.image.load('gameimages/candy4.jpg')
candy_image5 = pygame.image.load('gameimages/candy5.jpg')
#보너스 이미지
bonus_image = pygame.image.load('gameimages/bonus.jpg')

# 플레이어 초기화
player_width = player_image.get_size()[0]
player_height = player_image.get_size()[1]
player_pos_x = SCREEN_WIDTH / 2 - player_width / 2
player_pos_y = SCREEN_HEIGHT - player_height

# 호박 초기화
pumpkin1_width = pumpkin_image1.get_size()[0]
pumpkin1_height = pumpkin_image1.get_size()[1]
pumpkin1_pos_x = random.randint(0, SCREEN_WIDTH - pumpkin1_width)
pumpkin1_pos_y = -random.randint(pumpkin1_height, 300)

pumpkin2_width = pumpkin_image2.get_size()[0]
pumpkin2_height = pumpkin_image2.get_size()[1]
pumpkin2_pos_x = random.randint(0, SCREEN_WIDTH - pumpkin2_width)
pumpkin2_pos_y = -random.randint(pumpkin2_height, 300)

pumpkin3_width = pumpkin_image3.get_size()[0]
pumpkin3_height = pumpkin_image3.get_size()[1]
pumpkin3_pos_x = random.randint(0, SCREEN_WIDTH - pumpkin3_width)
pumpkin3_pos_y = -random.randint(pumpkin3_height, 300)

pumpkin4_width = pumpkin_image4.get_size()[0]
pumpkin4_height = pumpkin_image4.get_size()[1]
pumpkin4_pos_x = random.randint(0, SCREEN_WIDTH - pumpkin4_width)
pumpkin4_pos_y = -random.randint(pumpkin4_height, 300)

pumpkin5_width = pumpkin_image5.get_size()[0]
pumpkin5_height = pumpkin_image5.get_size()[1]
pumpkin5_pos_x = random.randint(0, SCREEN_WIDTH - pumpkin5_width)
pumpkin5_pos_y = -random.randint(pumpkin5_height, 500)

# 사탕 초기화
candy1_width = candy_image1.get_size()[0]
candy1_height = candy_image1.get_size()[1]
candy1_pos_x = random.randint(0, SCREEN_WIDTH - candy1_width)
candy1_pos_y = -random.randint(candy1_height, 300)

candy2_width = candy_image2.get_size()[0]
candy2_height = candy_image2.get_size()[1]
candy2_pos_x = random.randint(0, SCREEN_WIDTH - candy2_width)
candy2_pos_y = -random.randint(candy2_height, 300)

candy3_width = candy_image3.get_size()[0]
candy3_height = candy_image3.get_size()[1]
candy3_pos_x = random.randint(0, SCREEN_WIDTH - candy3_width)
candy3_pos_y = -random.randint(candy3_height, 300)

candy4_width = candy_image4.get_size()[0]
candy4_height = candy_image4.get_size()[1]
candy4_pos_x = random.randint(0, SCREEN_WIDTH - candy4_width)
candy4_pos_y = -random.randint(candy4_height, 300)

candy5_width = candy_image5.get_size()[0]
candy5_height = candy_image5.get_size()[1]
candy5_pos_x = random.randint(0, SCREEN_WIDTH - candy5_width)
candy5_pos_y = -random.randint(candy5_height, 300)

#보너스 객체 생성
bonus_width = bonus_image.get_size()[0]
bonus_height = bonus_image.get_size()[1]
bonus_pos_x = random.randint(0, SCREEN_WIDTH - bonus_width)
bonus_pos_y = -random.randint(bonus_height, 300)

# 시계 객체 생성
clock = pygame.time.Clock()

# 게임 루프를 제어하는 변수
is_running = True

# 목숨 설정
life = 5

# 스코어 설정
score = 0

# 게임 루프 시작
while is_running:
    # FPS 설정
    clock.tick(60)
    


    # X 버튼을 누르면 게임 종료
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            is_running = False

    # 키보드 입력 신호 받아옴
    key = pygame.key.get_pressed()

    # 왼쪽 방향키 -> 왼쪽으로 이동
    if key[pygame.K_LEFT] == 1:
        player_pos_x -= 700 * clock.get_time() / 1000

    # 오른쪽 방향키 -> 오른쪽으로 이동
    if key[pygame.K_RIGHT] == 1:
        player_pos_x += 700 * clock.get_time() / 1000

    # 좌측 맨 끝과 우측 맨 끝을 벗어나지 못하게 설정
    if player_pos_x < 0:
        player_pos_x = 0

    if player_pos_x > SCREEN_WIDTH - player_width:
        player_pos_x = SCREEN_WIDTH - player_width

    # 호박이 떨어지게 설정
    pumpkin1_pos_y += 100 * clock.get_time() / 1000
    pumpkin2_pos_y += 300 * clock.get_time() / 1000
    pumpkin3_pos_y += 500 * clock.get_time() / 1000
    pumpkin4_pos_y += 700 * clock.get_time() / 1000
    pumpkin5_pos_y += 900 * clock.get_time() / 1000
    
    # 사탕이 떨어지게 설정
    candy1_pos_y += 100 * clock.get_time() / 1000
    candy2_pos_y += 300 * clock.get_time() / 1000
    candy3_pos_y += 500 * clock.get_time() / 1000
    candy4_pos_y += 700 * clock.get_time() / 1000
    candy5_pos_y += 900 * clock.get_time() / 1000
    
    #보너스가 떨어지게 설정
    bonus_pos_y += 1000 * clock.get_time() / 1000

    # 호박이 바닥에 닿으면 랜덤한 위치에 다시 생성
    if pumpkin1_pos_y > SCREEN_HEIGHT - pumpkin1_height:
        pumpkin1_pos_x = random.randint(0, SCREEN_WIDTH - pumpkin1_width)
        pumpkin1_pos_y = -random.randint(pumpkin1_height, 300)
    
    if pumpkin2_pos_y > SCREEN_HEIGHT - pumpkin2_height:
        pumpkin2_pos_x = random.randint(0, SCREEN_WIDTH - pumpkin2_width)
        pumpkin2_pos_y = -random.randint(pumpkin2_height, 300)
        
    if pumpkin3_pos_y > SCREEN_HEIGHT - pumpkin3_height:
        pumpkin3_pos_x = random.randint(0, SCREEN_WIDTH - pumpkin3_width)
        pumpkin3_pos_y = -random.randint(pumpkin3_height, 300)
        
    if pumpkin4_pos_y > SCREEN_HEIGHT - pumpkin4_height:
        pumpkin4_pos_x = random.randint(0, SCREEN_WIDTH - pumpkin4_width)
        pumpkin4_pos_y = -random.randint(pumpkin4_height, 300)
        
    if pumpkin5_pos_y > SCREEN_HEIGHT - pumpkin5_height:
        pumpkin5_pos_x = random.randint(0, SCREEN_WIDTH - pumpkin5_width)
        pumpkin5_pos_y = -random.randint(pumpkin5_height, 500)
    
    # 사탕이 바닥에 닿으면 랜덤한 위치에 다시 생성
    if candy1_pos_y > SCREEN_HEIGHT - candy1_height:
        candy1_pos_x = random.randint(0, SCREEN_WIDTH - candy1_width)
        candy1_pos_y = -random.randint(candy1_height, 300)
        
    if candy2_pos_y > SCREEN_HEIGHT - candy2_height:
        candy2_pos_x = random.randint(0, SCREEN_WIDTH - candy2_width)
        candy2_pos_y = -random.randint(candy2_height, 300)
        
    if candy3_pos_y > SCREEN_HEIGHT - candy3_height:
        candy3_pos_x = random.randint(0, SCREEN_WIDTH - candy3_width)
        candy3_pos_y = -random.randint(candy3_height, 300)
        
    if candy4_pos_y > SCREEN_HEIGHT - candy4_height:
        candy4_pos_x = random.randint(0, SCREEN_WIDTH - candy4_width)
        candy4_pos_y = -random.randint(candy4_height, 300)
        
    if candy5_pos_y > SCREEN_HEIGHT - candy5_height:
        candy5_pos_x = random.randint(0, SCREEN_WIDTH - candy5_width)
        candy5_pos_y = -random.randint(candy5_height, 300)
        
    #보너스가 바닥에 닿으면 랜덤한 위치에 다시 생성
    if bonus_pos_y > SCREEN_HEIGHT - bonus_height:
        bonus_pos_x = random.randint(0, SCREEN_WIDTH - bonus_width)
        bonus_pos_y = -random.randint(bonus_height, 300)
        
        

    # 플레이어의 렉트 정보 업데이트
    player_rect = player_image.get_rect()
    player_rect.x = player_pos_x
    player_rect.y = player_pos_y

    # 호박의 렉트 정보 업데이트
    pumpkin1_rect = pumpkin_image1.get_rect()
    pumpkin1_rect.x = pumpkin1_pos_x
    pumpkin1_rect.y = pumpkin1_pos_y
    
    pumpkin2_rect = pumpkin_image2.get_rect()
    pumpkin2_rect.x = pumpkin2_pos_x
    pumpkin2_rect.y = pumpkin2_pos_y
    
    pumpkin3_rect = pumpkin_image3.get_rect()
    pumpkin3_rect.x = pumpkin3_pos_x
    pumpkin3_rect.y = pumpkin3_pos_y
    
    pumpkin4_rect = pumpkin_image4.get_rect()
    pumpkin4_rect.x = pumpkin4_pos_x
    pumpkin4_rect.y = pumpkin4_pos_y
    
    pumpkin5_rect = pumpkin_image5.get_rect()
    pumpkin5_rect.x = pumpkin5_pos_x
    pumpkin5_rect.y = pumpkin5_pos_y
    
    #사탕의 렉트 정보 업데이트
    candy1_rect = candy_image1.get_rect()
    candy1_rect.x = candy1_pos_x
    candy1_rect.y = candy1_pos_y
    
    candy2_rect = candy_image2.get_rect()
    candy2_rect.x = candy2_pos_x
    candy2_rect.y = candy2_pos_y
    
    candy3_rect = candy_image3.get_rect()
    candy3_rect.x = candy3_pos_x
    candy3_rect.y = candy3_pos_y
    
    candy4_rect = candy_image4.get_rect()
    candy4_rect.x = candy4_pos_x
    candy4_rect.y = candy4_pos_y
    
    candy5_rect = candy_image5.get_rect()
    candy5_rect.x = candy5_pos_x
    candy5_rect.y = candy5_pos_y
    
    #보너스의 렉트 정보 업데이트
    bonus_rect = bonus_image.get_rect()
    bonus_rect.x = bonus_pos_x
    bonus_rect.y = bonus_pos_y

    # 호박 충돌 검사
    if player_rect.colliderect(pumpkin1_rect):
        life -= 1
        if life == 0:
            is_running = False
            
    if player_rect.colliderect(pumpkin2_rect):
        life -= 1
        if life == 0:
            is_running = False
            
    if player_rect.colliderect(pumpkin3_rect):
        life -= 1
        if life == 0:
            is_running = False
            
    if player_rect.colliderect(pumpkin4_rect):
        life -= 1
        if life == 0:
            is_running = False
            
    if player_rect.colliderect(pumpkin5_rect):
        life -= 1
        if life == 0:
            is_running = False
            
    # 사탕 충돌 검사
    if player_rect.colliderect(candy1_rect):
        score += 10
    if player_rect.colliderect(candy2_rect):
        score += 30
    if player_rect.colliderect(candy3_rect):
        score += 50
    if player_rect.colliderect(candy4_rect):
        score += 70
    if player_rect.colliderect(candy5_rect):
        score += 90
    
    #보너스 충돌 검사
    if player_rect.colliderect(bonus_rect):
        print('보너스')

    # 렌더(그리기)
    screen.blit(background_image, (0, 0))
    screen.blit(player_image, (player_pos_x, player_pos_y))
    screen.blit(pumpkin_image1, (pumpkin1_pos_x, pumpkin1_pos_y))
    screen.blit(pumpkin_image2, (pumpkin2_pos_x, pumpkin2_pos_y))
    screen.blit(pumpkin_image3, (pumpkin3_pos_x, pumpkin3_pos_y))
    screen.blit(pumpkin_image4, (pumpkin4_pos_x, pumpkin4_pos_y))
    screen.blit(pumpkin_image5, (pumpkin5_pos_x, pumpkin5_pos_y))
    screen.blit(candy_image1, (candy1_pos_x, candy1_pos_y))
    screen.blit(candy_image2, (candy2_pos_x, candy2_pos_y))
    screen.blit(candy_image3, (candy3_pos_x, candy3_pos_y))
    screen.blit(candy_image4, (candy4_pos_x, candy4_pos_y))
    screen.blit(candy_image5, (candy5_pos_x, candy5_pos_y))
    screen.blit(bonus_image, (bonus_pos_x, bonus_pos_y))

    # 디스플레이 업데이트(필수)
    pygame.display.update()

# while문을 탈출하면 pygame 종료
pygame.quit()

```
