### 단어 빈도 계산

```python
from konlpy.tag import Okt
from collections import Counter
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ratings.txt', delimiter='\t')
df
#text파일이지만 read_csv로 불러와진다.
#csv파일은 ,로 나누어지지만 이 파일은 열을 tab으로 구분한다. 
#그 때문에 delimiter속성을 \t로 바꾸어주면 정상적으로 나뉘는 것을 볼 수 있다.
```

```python
x = df['document'][:3000]
#x는 df의 document열을 3000개만 사용
x
```

```python
n = []

#리뷰들을 리스트화
for i in x:
    n.append(i)
n
```

```python
okt = Okt()
#Okt 형태소분석기 사용
noun = okt.nouns(str(n))
#명사 추출, n은 리스트기 때문에 형태소 추출을 위해 str로 변환
```

```python
#명사 빈도 확인
count = Counter(noun)
```

```python
count
```

```python
#명사 빈도 카운트
noun_list = count.most_common(100) #많이 나온 100개 단어
```

```python
for i in noun_list:
    print(i)
```

### 단어 리스트 만들기

```python
word_list = []
word_list = df['document'][:3000]
```

```python
t = []
for i in word_list:
    morph = okt.pos(i)
    t.append(morph)
```

### 명사 추출

```python
noun_list = []
for s in t:
    for word,tag in s:
        if tag in ['Noun']:
            noun_list.append(word)
```

```python
noun_list
```

### 두 글자 이상 단어만 추출

```python
 noun_list2 = [n for n in noun_list if len(n)>1]
```

```python
 noun_list2 
```

### 단어별 개수 세기 

```python
counts = Counter(noun_list2)
```

```python
counts
```

```python
tags =counts.most_common(200)
```

```python
tags
```

```python
dic = dict(tags)
#dict(a)= a를 딕셔너리화하는 함수
```
