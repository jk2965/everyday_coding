## 1.목표 - 현대차 가격 예측하기

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.ensemble import RandomForestRegressor
```

## 2.데이터 수집 

```python
train_df = pd.read_excel('현대차_가격.xlsx',sheet_name='train')
test_df = pd.read_excel('현대차_가격.xlsx',sheet_name='test')
```

```python
train_df
```

```python
test_df
```

```python
x_train = train_df.drop(['가격'],axis=1)
x_test = test_df.drop(['가격'],axis=1)
y_train = train_df['가격']
y_test = test_df['가격']
#데이터 분할
```

```python
x_train
```

```python
x_test
```

### 3.데이터 인코딩

#### One_hot encoding

```python
transformer = make_column_transformer((OneHotEncoder(),
                                       ['종류','연료','변속기']))
#One hot 인코딩을 위해 transformer 모델을
#make_column_transformer((OneHotEncoder(),[바꾸려고하는 칼럼명]))
#으로 변환
```

```python
transformer.fit(x_train)
#모델 학습
```

```python
x_train = transformer.transform(x_train)
x_test = transformer.transform(x_test)
#transformer 모델의 transform 내장함수로 x_train과 x_test를 변경시킴
```

```python
x_train
```

## 모델생성

### random Forest

```python
rf = RandomForestRegressor(n_estimators=1000,max_depth=3,
                           max_leaf_nodes=10).fit(x_train,y_train)
#모델 선정, 1000개의 트리, 깊이 = 3, 
```

```python
pred = rf.predict(x_test)
pred
#예측값 산출
```

```python
rf.score(x_train,y_train)
#회귀 모델의 model.score()의 기준 = r2_score
```

```python
rf.score(x_test,y_test)
#회귀 모델의 model.score()의 기준 = r2_score
```

```python
new_test=[[2015,'대형',13.3,170,24.0,'LPG',0,2500,2000,'수동']]
new_test=pd.DataFrame(new_test,columns=['년식','종류','연비','마력',
                                        '토크','연료','하이브리드',
                                        '배기량','중량','변속기'] )
#새로운 데이터 생성
```

```python
new_test=transformer.transform(new_test)

#새로운 데이터 One-hot Encoding
```

```python
y_predict = rf.predict(new_test)
y_predict
#예측값 산출
```
